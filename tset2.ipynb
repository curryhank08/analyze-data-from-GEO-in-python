```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Create an empty DataFrame to store the results
results_df = pd.DataFrame(columns=['Probe_ID', 'P-value_Age', 'P-value_Gender', 'R-squared'])

# Loop through each probe in the pivoted_samples DataFrame
for probe_id in pivoted_samples.columns:
    # Merge Age, Gender, and Expression data
    probe_data = pd.concat([sample_age_gender_data, pivoted_samples[probe_id]], axis=1)
    
    # Perform linear regression using scikit-learn
    X = probe_data[['Age', 'Gender']]
    y = probe_data[probe_id]
    
    # Fit the linear regression model
    model = LinearRegression().fit(X, y)
    
    # Calculate R-squared
    y_pred = model.predict(X)
    r_squared = r2_score(y, y_pred)
    
    # Perform ordinary least squares (OLS) regression to get p-values
    X = sm.add_constant(X)  # Add a constant term (intercept)
    model_ols = sm.OLS(y, X).fit()
    
    # Extract p-values
    p_values = model_ols.pvalues[1:]  # Exclude the constant term
    
    # Append results to the results DataFrame
    results_df = results_df.append({'Probe_ID': probe_id, 'P-value_Age': p_values['Age'], 'P-value_Gender': p_values['Gender'], 'R-squared': r_squared}, ignore_index=True)

# Print the results DataFrame
print(results_df.head())
```

```python
import json
import requests

ENRICHR_URL = 'https://amp.pharm.mssm.edu/Enrichr'

def _enrichr_add_list(genes, meta=''):
    genes_str = '\n'.join(genes)
    payload = {
        'list': (None, genes_str),
        'description': (None, meta)
    }
    # POST genes to the /addList endpoint
    response = requests.post("%s/addList" % ENRICHR_URL, files=payload)
    list_ids = json.loads(response.text)
    return list_ids

def enrichr_link(genes, meta=''):
    list_ids = _enrichr_add_list(genes, meta)
    shortId = list_ids['shortId']
    link = '%s/enrich?dataset=%s' % (ENRICHR_URL, shortId)
    return link
```
